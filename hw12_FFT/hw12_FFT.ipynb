```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft,fftshift
```

```python
L = 16
N = 256
dx = 2 * L / N
x =  np.linspace(-L, L, N, endpoint=False)
```

```python
def bump_func(x, d):
    return np.where(np.abs(x) < np.abs(d), d**2 - x**2, 0.0)
```

```python
f_x = bump_func(x, 2) - bump_func(x, 4)/8
```

```python
plt.figure(figsize=(5, 3))
plt.plot(x, f_x)
plt.xlim(-L, L)
plt.title("Function in x-space")
plt.xlabel("x")
plt.ylabel("f(x)")
```

#### FFT Processing

```python
'''
1) Shift x=0 to the initial index before Transformation
2) Perform FFT
3) Shift k=0 back to the center of the array
'''
f_k = fftshift(fft(fftshift(dx*f_x)))

K = np.pi * N / (2 * L)
k = np.linspace(-K, K, N, endpoint=False)
```

```python
#plt.figure(figsize=(5, 3))
plt.plot(k, f_k)
plt.xlim(-K, K)
plt.title("FFT in k-space")
plt.xlabel("k")
plt.ylabel("F(k)")
```
