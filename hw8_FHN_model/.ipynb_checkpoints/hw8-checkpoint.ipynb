```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler_step(t,y,f,dt):
    dyt = f(t,y)
    return y+dt*dyt
```

```python
class FHN_model:
    def __init__(self,mr = 50 ,ie = 0, fa = 0.66, fb = 0.5, tw = 10):
        self.mr = mr
        self.ie = ie
        self.fa = fa
        self.fb = fb
        self.tw = tw

    def fhn_slope(self, t,x):
        return np.array([
        x[0]-x[0]**3/3-x[1]+self.mr*self.ie,
        (x[0]+self.fa-self.fb*x[1])/self.tw
    ])

    def get_traces(self, tt,dt,xi,ies=None,itg=euler_step):
        global ie
        nf = int(tt/dt)
        if ies is None: ies = np.zeros(nf)
        ti = 0
        x = xi
        # Traces
        tis = []   # Time frames
        xs = []    # Traces
        for i in range(nf):
            ie = ies[i%len(ies)]
            #print(x)
            x = itg(ti,x,self.fhn_slope,dt)
            ti += dt
            tis.append(ti)
            xs.append(x.copy())
        return np.array(tis),np.array(xs)
```

```python
fhn = FHN_model()

ts,xs = fhn.get_traces(300.0,0.2,np.array([1.,0.]))
```

```python
plt.plot(ts,xs[:,0],label='$v$')
plt.plot(ts,xs[:,1],label='$w$')
plt.xlabel('Time')
plt.xlim(0,300.0)
plt.legend()
plt.show()
```

### Q2

```python
import brian2 as b2
```

```python
mr = 50
ie = 0
fa = 0.66
fb = 0.5
tw = 10
```

```python
if 'ms' in globals(): del ms
```

```python
tw = 10 * b2.ms  # 
dt = 0.2 * b2.ms
tt = 300 * b2.ms
pulse_start = 0 * b2.ms
pulse_end = 300 * b2.ms
I_pulse_amp = 0 * b2.uA  

#  (TimedArray)
num_steps = int(tt / dt)
current_array = np.zeros(num_steps) * b2.uA
current_array[int(pulse_start/dt):int(pulse_end/dt)] = I_pulse_amp


#  Brian2  TimedArray
input_current = b2.TimedArray(current_array, dt=dt)


eqs = """
ie = input_current(t) / uA :1
dv/dt = (v - v**3/3 - w + ie*mr) / ms : 1
dw/dt = (v + fa - fb*w)/tw : 1
"""

#  (NeuronGroup) 
#  Euler FHN  ODE
neuron = b2.NeuronGroup(1, eqs, method='euler', 
                         namespace={'fa': fa, 'fb': fb, 'tw': tw, 'mr': mr, 
                                    'input_current': input_current})

# 
neuron.v = 1.0 
neuron.w = 0.0 

#  v, w
mon = b2.StateMonitor(neuron, ['v', 'w'], record=True)

b2.run(duration=tt)

plt.plot(mon.t/b2.ms, mon.v[0], label='v')
plt.plot(mon.t/b2.ms, mon.w[0], label='w')

plt.xlim(0, 300)
plt.legend()
plt.show()
```
